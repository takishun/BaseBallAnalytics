```python
data = pd.DataFrame()
```

```python
teams_b=['b','c','d','db','e','f','g','h','l','m','s','t']
teams_s=['bs','c','d','e','f','h','l','m','s','t','g','yb']
```

```python
b_data = 'batter_data_2019_'
s_data = 'output_'
```

```python
# predictions = pd.concat([predictions,pd.DataFrame(y_pred,columns=['pred'+str(i)])],axis=1)
batting_data=pd.DataFrame()
profile_data=pd.DataFrame()
length = 0
for i in teams_b:
    filename = 'data/' + b_data + str(i) + '.csv'
    one_time_data=pd.read_csv(filename)
    batting_data=pd.concat([batting_data,pd.DataFrame(one_time_data,columns=one_time_data.columns)])

for i in teams_s:
    filename = 'data/' + s_data + str(i) + '.csv'
    one_time_data=pd.read_csv(filename)
    print(filename)
    profile_data=pd.concat([profile_data,pd.DataFrame(one_time_data,columns=one_time_data.columns)])
    length += len(one_time_data)

print(length)        
```

```python
batting_data
```

```python
profile_data
```

```python
merged_data = pd.merge(batting_data,profile_data,on='選手名',how='inner')
```

```python
batter_data = merged_data[merged_data['守備'] != '投手']
```

```python
batter_data.reset_index(drop=True,inplace=True)
```

```python
batter_data.to_csv('data/all_batter_data.csv')
```

```python
batter_data['BMI']=batter_data.loc[:,'体重']/((batter_data.loc[:,'身長']/100)**2)
```

```python
batter_data['BMI']=batter_data['BMI'].round(2)
```

```python
batter_data.to_csv('data/all_batter_data.csv',index=False)
```
